# 3D plot of A3 Transformation Temperatures

```python
#importing the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
```

```python
#preparing the Data
df = pd.read_csv("plot points ordered - Sheet1.csv")
#Dropping all NaN values in the Relaxed condition: Temperature column
df.dropna( subset = ['Relaxed condition: Temperature'], inplace = True)
```

```python
#Now let us assign each column into an axis and make an array
#Let us typecast the df column into a list to remove the index and then put it into the list
x_data = [list(df['Mass percent Al'])]
y_data = [list(df['Mass percent Mn'])]
z_data = [list(df['Mass percent Si'])]
#an array of temperatures that will be used to define the color later
colors = [list(df['Relaxed condition: Temperature'])]
```

```python
#let us use colormap to bring the varying color in
from matplotlib.colors import ListedColormap

#creating a 3D axis
axis = plt.axes(projection = "3d")

#shifting the view
axis.view_init(elev=30, azim=-120)

#creating the plot and mapping color to temperature
scatter = axis.scatter(x_data, y_data, z_data, c = colors, cmap = 'Reds', alpha = 0.8)
plt.colorbar(scatter, ax=axis)
axis.set_xlabel("Mass percent Al")
axis.set_ylabel("Mass percent Mn")
axis.set_zlabel("Mass percent Si")
plt.show()
```
